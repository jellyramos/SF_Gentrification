```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
file = "SF_foodstoreData.csv"

foodstores_raw = pd.read_csv(file)
foodstores_raw
```

```python
del foodstores_raw["Unnamed: 14"]
foodstores_raw
```

```python
foodstores_df = foodstores_raw.drop(foodstores_raw.index[[312,313,314,315,316,317]])
foodstores_df
```

```python
foodstores_df.dtypes
```

```python
x_var = "Post Code"
groupby_var = "Category Labels"
df_agg = foodstores_df.loc[:, [x_var, groupby_var]].groupby(groupby_var)
vals = [foodstores_df[x_var].values.tolist() for i, foodstores_df in df_agg]
```

```python
plt.figure(figsize=(16,9), dpi= 80)
colors = [plt.cm.Spectral(i/float(len(vals)-1)) for i in range(len(vals))]
n, bins, patches = plt.hist(vals, foodstores_df[x_var].unique().__len__(), stacked=True, density=False, color=colors[:len(vals)])

plt.legend({group:col for group, col in zip(np.unique(foodstores_df[groupby_var]).tolist(), colors[:len(vals)])}, loc='upper right')
plt.title("Food Store Types in San Franciscso by Post Code", fontsize=22)
plt.xlabel(x_var)
plt.ylabel("Number of Stores")
plt.ylim(0, 55)
plt.xticks(ticks=bins, labels=np.unique(foodstores_df[x_var]).tolist(), rotation=45, horizontalalignment='left')
plt.show()
```
