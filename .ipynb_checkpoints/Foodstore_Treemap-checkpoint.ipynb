```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import squarify
```

```python
file = "SF_foodstoreData.csv"

foodstores_df = pd.read_csv(file)
foodstores_df
```

```python
foodcounts_df = foodstores_df.groupby("Category Labels").size().reset_index(name="Counts")
labels = foodcounts_df.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1)
sizes = foodcounts_df["Counts"].values.tolist()
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

plt.figure(figsize=(12,8), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

plt.title('Treemap of Store Types in San Francisco', fontsize=22)
plt.axis('off')
plt.show()
```
