```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "SF_foodstoreData.csv"

foodstores_raw = pd.read_csv(file)
foodstores_raw
```

```python
del foodstores_raw["Unnamed: 14"]
foodstores_raw
```

```python
foodstores_df = foodstores_raw.drop(foodstores_raw.index[[312,313,314,315,316,317]])
foodstores_df
```

```python
foodstores_df.dtypes
```

```python
value_counts = foodstores_df["Post Code"].value_counts()
print(value_counts)
```

```python
postcode_count = pd.DataFrame(value_counts)
postcode_count
```

```python
postcode_count = postcode_count.reset_index()
postcode_count.columns = ["Post Code", "No. of Stores in Area"]
postcode_count
```

```python
x_axis = np.arange(len(postcode_count))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(16,9))
plt.bar(x_axis, postcode_count["No. of Stores in Area"], align="edge", 
        color=("#5964ff", 
               "#6f59ff", 
               "#8859ff", 
               "#8859ff", 
               "#aa59ff", 
               "#b259ff", "#b259ff", "#b259ff", 
               "#c359ff", 
               "#cb59ff", "#cb59ff",
               "#e459ff", "#e459ff", 
               "#ed59ff", "#ed59ff", 
               "#ff59f8", "#ff59f8", 
               "#ff59e7", "#ff59e7", 
               "#ff59d7", 
               "#ff59c6", 
               "#ff59a4", "#ff59a4", 
               "#ff5994", 
               "#ff5961", "#ff5961"))
plt.xticks(tick_locations, postcode_count["Post Code"], rotation=45)

plt.title("Number of Food Stores per San Francisco Zip Code", fontsize=22)
plt.xlabel("Post Codes")
plt.ylabel("Number of Stores")
```
